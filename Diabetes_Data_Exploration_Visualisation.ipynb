# **DATA VISUALISATION**

## 4.Utilizing the pandas library's pd.read_csv function, read the dataset

Checking type and info about the data in dataset using dtypes and info method

```python
df = pd.read_csv("Diabetes0.csv")
df
```

## 5.Display the target variable's distribution graphically.
A bar plot of the target variable's distribution (in this case, "class") is produced using sns.countplot. This graphic sheds light on how well or poorly the classes are balanced.

Understanding the distribution of the target variable—in this case, "class"—is made easier by this depiction.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=df)
plt.title('Distribution of Classes')
plt.show()
```

```python
X = df.drop('class', axis=1)
y = df['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='bar')
plt.title('Feature Importances')
plt.show()
```

## 6.Visualize the distribution of numeric features

*  sns.boxplot:Generates a boxplot for each numeric feature.
*  You can use a boxplot to show the distribution, central tendency, and dispersion of each numeric attribute in your DataFrame. The boxplot displays data on the median, quartiles, and probable outliers.

```python

# Box plot
sns.boxplot(data=df[['Age', 'Pregnancies']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=df[['PG_Concentration','Diastolic_BP']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=df[['Serum_Ins','BMI','DP_Function']])
plt.title('Box Plot of Age and feature2')
plt.show()
```

The same can be done using histograms also .The image shows the distribution of each numerical property, revealing their primary tendencies and spread.

```python
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_features].hist(figsize=(12, 10))
plt.suptitle('Histograms of Numeric Features', x=0.5, y=1.02, ha='center', fontsize='large')
plt.show()
```

##7. Visualizing pairwise relationships in a dataset

>
A **scatter plot **displays the relationship between two numerical variables visually. Individual data points are represented by dots, and additional dimensions may be represented by indicators such as color or form.


```python
sns.scatterplot(x='Age', y='Tri_Fold_Thick', data= df)
plt.title('Scatter Plot of Age vs Tri_Fold_Thick')
plt.show()
```

## 8.HISTOGRAM
Histograms show similarities, changes, and distribution patterns across several datasets. They enable a comparison of the forms and variances within the data distributions by shedding light on similarities, central tendencies, and differences.
Here the overlapping and  comparison between Tri Fold Thick and Body MAss Index is compared and visualised graphically

```python
sns.histplot(data=df, x='Tri_Fold_Thick', bins=20, color='blue', alpha=0.9, label='Tri_Fold_Thick')
sns.histplot(data=df, x='BMI', bins=22, color='orange', alpha=0.9, label='BMI')
plt.title('Histogram of Tri_Fold_Thick and BMI')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```


## 9.Visualization using the correlation matrix

A heatmap can be used to visualize dataset correlations, demonstrating links between variables. It can help you understand patterns and dependencies.



The correlation matrix for numeric features is computed using df.corr().

sns.heatmap creates a heatmap of the correlation matrix, revealing the correlations between various numeric features.





```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## 10.Temporal trends and patterns analysis.

**Time series** graphs depict the evolution of a variable across time, exhibiting trends, patterns, and probable future values.

## 11.Visualising using Pair Plot
Pairplot displays scatter plots for numerical variables and distributions for univariate analysis in a dataset to aid in comprehensive data exploration.
sns.pairplot generates a pairplot for a set of features, displaying scatterplots for each pair of features and coloring the points according to the target variable ('class').


This image aids in comprehending the links between features and their distribution.



```python
selected_features = ['Pregnancies', 'PG_Concentration', 'Diastolic_BP', 'class']
sns.pairplot(df[selected_features], hue='class', markers=['o', 's'])
plt.suptitle('Pairplot of Selected Features', y=1.02, ha='center', fontsize='large')
plt.show()
```
